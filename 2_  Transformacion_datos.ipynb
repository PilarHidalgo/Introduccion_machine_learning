# **Transformar datos categ칩ricos**


La mayor칤a de los algoritmos de machine learning funcionan con datos num칠ricos. Pero algunos conjuntos de datos tambi칠n contienen datos categ칩ricos, para aplicar algoritmos de aprendizaje autom치tico necesitamos convertir caracter칤sticas categ칩ricas en representaci칩n num칠rica, esta tarea se denomina *encoding*. En este sentido, consideramos dos tipos de escala categ칩rica, nominal y ordinal.

## **Escala nominal**: 
La escala nominal se refiere a variables que solo se nombran. Estos nombres son etiquetas y su valor no tiene significado num칠rico, es decir, las operaciones matem치ticas b치sicas no tienen ning칰n sentido. Por ejemplo, si **Sexo:** mujer=1 y hombre=2; mujer+hombre no tiene sentido.

## **Escala ordinal**: 
Una escala ordinal es cuando una variable tiene un valor con significado correspondiente a una etiqueta. Por ejemplo, una escala de Likert para la calidad de los productos: Mala=1, Regular=2, Buena=3, Excelente=4.

Cada tipo de escala tiene diferentes estrategias para codificar. Vamos a explicar 4 tipos: Codificaci칩n de etiquetas, Codificaci칩n One-hot, Codificaci칩n ordinal y Codificaci칩n binaria.

## **Estrategias para escala ordinal**:

````python
#Vamos por el dataset de autos
import pandas as pd
import numpy as np
# Definir las cabeceras porque este data set no tiene
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]
# Leer desde la pagina de UCI
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data",
                  header=None, names=headers, na_values="?" )
df.head()
````


## 1.   **Encontrar y reemplazar:** 
Si los datos contienen nombres como "one", "two", "three"; es posible cambiar los valores categ칩ricos a n칰meros como 1,2,3. Por ejemplo las columnas "num_puertas" y "num_cilindros".

````python
# Hacemos el diccionario:
cleanup_nums = {"num_doors":     {"four": 4, "two": 2},
                "num_cylinders": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "twelve": 12, "three":3 }}
````

````python
#Usamos el comando "replace"
obj_df = df.select_dtypes(include=['object']).copy()
obj_df = obj_df.replace(cleanup_nums)
obj_df.head()
````


## 2.   **Label encoding:** 
Para convertir categor칤as en n칰meros, podemos asignar valores num칠ricos autom치ticamente. Por ejemplo, la columna "body_style".

````python
#Convertir los datos de columna en categorias
obj_df["body_style"] = obj_df["body_style"].astype('category')
#Para saber cu치ntas categor칤as existen, usamos count_values
obj_df["body_style"].value_counts()
````

````python
#Con el comando "cat.codes" convertimos los nombres a valores
obj_df["body_style_cat"] = obj_df["body_style"].cat.codes
#Tenemos 5 categorias 
obj_df["body_style_cat"].value_counts()
````

````python
#Podemos usar tambien sklearn
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
obj_df["make_code"] = ord_enc.fit_transform(obj_df[["make"]])
obj_df[["make", "make_code"]].value_counts()
````

No importa cu치ntos registros tenga cada categor칤a, el algoritmo le asigna un valor aleatorio y eso podr칤a ser un problema. En algunos casos, es importante el valor que le vamos a dar a la categor칤a, por lo que si consideramos que alguna categor칤a merece una puntuaci칩n o no, tenemos que fijarnos en qu칠 valor le vamos a dar. En la variable "body_style", en realidad no tiene sentido poner valores de ranking ya que "sed치n (n칰mero 3) no es mejor que "wagon" (n칰mero 4), al menos que tengas un criterio para eso. Si no lo tienes Si no hay raz칩n para asignar este ranking, es mejor usar otra estrategia, ya que estas variables son nominales y no ordinales.

## 3.   **One hot encoding:** 
En lugar de asignar un valor a cada categor칤a, podemos agregar columnas seg칰n cada categor칤a, de esta manera todas las categor칤as tienen el mismo peso y el modelo no se considerar치 sobre otras debido al valor.

````python
#Podemos usar get_dummies para hacer las otras columnas, una biblioteca de pandas
pd.get_dummies(obj_df, columns=["body_style", "drive_wheels"], prefix=["body", "drive"]).head()
````

````python
#Tambien con sklearn
from sklearn.preprocessing import OneHotEncoder
oe_style = OneHotEncoder()
oe_results = oe_style.fit_transform(obj_df[["body_style"]])
pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_).head()
````

## 4.   **Custom binary encoding:** 

El mismo principio de codificaci칩n One-hot, puede usar numpy para hacer *dummies*

````python
obj_df["OHC_Code"] = np.where(obj_df["engine_type"].str.contains("ohc"), 1, 0)
obj_df[["make", "engine_type", "OHC_Code"]].head()
````

**쯈u칠 pasa si usamos una estrategia ordinal con variables nominales?**: 游뱂


Hagamos un ejemplo con el modelo de regresi칩n lineal.



````python
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
# para los fines de este an치lisis, solo use un peque침o subconjunto de caracter칤sticas

feature_cols = [
    'fuel_type', 'make', 'aspiration', 'highway_mpg', 'city_mpg',
    'curb_weight', 'drive_wheels'
]

# Eliminar las filas de precios vac칤as
df_ml = df.dropna(subset=['price'])

X = df_ml[feature_cols]
y = df_ml['price']
````

````python
#Aqu칤 asignamos las variables nominales a valores de codificaci칩n one-hot para: fuel_type', 'make', 'drive_wheels', and Ordinal encoding for 'aspiration'
enc = make_column_transformer((OneHotEncoder(handle_unknown='ignore'),
                                        ['fuel_type', 'make', 'drive_wheels']),
                                      (OrdinalEncoder(), ['aspiration']),
                                      remainder='passthrough')
````

````python
X_transform = enc.fit_transform(X)

# Aplica la regresion lineal
model = LinearRegression()
model.fit(X_transform, y)
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_transform, y, test_size=0.33, random_state=42)
````

````python
model.fit(X_train,y_train)
````

````python
y_pred=model.predict(X_test)
````

````python
from sklearn import metrics
print(metrics.mean_absolute_percentage_error(y_test, y_pred))
````

````python
#Usando solo Ordinal encoder
enc = make_column_transformer((OrdinalEncoder(),
                                        ['fuel_type', 'make', 'drive_wheels']),
                                      (OrdinalEncoder(), ['aspiration']),
                                      remainder='passthrough')
````

````python
X_transform = enc.fit_transform(X)

# aplica regresi칩n lineal como quieras
model = LinearRegression()
model.fit(X_transform, y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_transform, y, test_size=0.33, random_state=42)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
````

````python
print(metrics.mean_absolute_percentage_error(y_test, y_pred))
````

Como podemos ver, si tratamos las variables nominales como ordinales, el porcentaje de error aumenta. En el ejemplo, el error con codificaci칩n nominal es 0,14 y con ordinal es 0,19.
